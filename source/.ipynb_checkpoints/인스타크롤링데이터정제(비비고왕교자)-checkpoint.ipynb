```python
import pandas as pd
from konlpy.tag import Twitter
from collections import Counter
import re
import pandas as pd
from wordcloud import WordCloud
```

```python
def get_tags(text, ntags=10):
    spliter = Twitter()
    # konlpyì˜ Twitterê°ì²´
    nouns = spliter.nouns(text)
    # nouns í•¨ìˆ˜ë¥¼ í†µí•´ì„œ textì—ì„œ ëª…ì‚¬ë§Œ ë¶„ë¦¬/ì¶”ì¶œ
    count = Counter(nouns)
    # Counterê°ì²´ë¥¼ ìƒì„±í•˜ê³  ì°¸ì¡°ë³€ìˆ˜ nounsí• ë‹¹
    return_list = []  # ëª…ì‚¬ ë¹ˆë„ìˆ˜ ì €ì¥í•  ë³€ìˆ˜
    for n, c in count.most_common(ntags):
        temp = {'tag': n, 'count': c}
        return_list.append(temp)
    # most_common ë©”ì†Œë“œëŠ” ì •ìˆ˜ë¥¼ ì…ë ¥ë°›ì•„ ê°ì²´ ì•ˆì˜ ëª…ì‚¬ì¤‘ ë¹ˆë„ìˆ˜
    # í° ëª…ì‚¬ë¶€í„° ìˆœì„œëŒ€ë¡œ ì…ë ¥ë°›ì€ ì •ìˆ˜ ê°¯ìˆ˜ë§Œí¼ ì €ì¥ë˜ì–´ìˆëŠ” ê°ì²´ ë°˜í™˜
    # ëª…ì‚¬ì™€ ì‚¬ìš©ëœ ê°¯ìˆ˜ë¥¼ return_listì— ì €ì¥í•©ë‹ˆë‹¤.
    return return_list 
```

```python
def main():
    text_file_name = "pandas.txt" #í•´ì‹œíƒœê·¸, ëŒ“ê¸€, ë³¸ë¬¸, í•©
    # ë¶„ì„í•  íŒŒì¼
    noun_count = 1000
    # ìµœëŒ€ ë§ì€ ë¹ˆë„ìˆ˜ ë¶€í„° 20ê°œ ëª…ì‚¬ ì¶”ì¶œ
    output_file_name = "cout.txt"
    # count.txt ì— ì €ì¥
    open_text_file = open(text_file_name, 'r',-1,"utf-8")
    # ë¶„ì„í•  íŒŒì¼ì„ open 
    text = open_text_file.read() #íŒŒì¼ì„ ì½ìŠµë‹ˆë‹¤.
    text = text.replace("\n","")
    text = re.sub('[-=+,#/\?:^$.@*\"â€»~&%ã†!ã€\\â€˜|\(\)\[\]\<\>`\'â€¦ã€‹]', '', text)
    hangul = re.compile('[^ ã„±-ã…£ê°€-í£]+')
    text = hangul.sub('', text) # í•œê¸€ê³¼ ë„ì–´ì“°ê¸°ë¥¼ ì œì™¸í•œ ëª¨ë“  ë¶€ë¶„ì„ ì œê±°
    tags = get_tags(text, noun_count) # get_tags í•¨ìˆ˜ ì‹¤í–‰
    open_text_file.close()   #íŒŒì¼ close
    open_output_file = open(output_file_name, 'w',-1,"utf-8")
    # ê²°ê³¼ë¡œ ì“°ì¼ count.txt ì—´ê¸°
    for tag in tags:
        noun = tag['tag']
        count = tag['count']
        open_output_file.write('{} {}\n'.format(noun, count))
    # ê²°ê³¼ ì €ì¥
    open_output_file.close() 
```
